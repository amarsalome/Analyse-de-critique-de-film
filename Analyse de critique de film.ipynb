Nettoyage du texte en retirant les stop-words. 
Création d'une nouvelle colonne contenant le texte nettoyé. 
On utilise nltk pour l'ensemble des stop words.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
stop_words = set(stopwords.words('english'))
```

```python
word_tokens = data_train["reviews"].apply(lambda x : " ".join([word for word in word_tokenize(str(x)) if word not in stop_words]))
```

```python
word_tokens
```

```python
data_train["new_reviews"] = word_tokens
data_train
```

```python

word_tokens2 = data_test["reviews"].apply(lambda x : " ".join([word for word in word_tokenize(str(x)) if word not in stop_words]))
```

```python
data_test["new_reviews"] = word_tokens2
```

```python
X_train = data_train.new_reviews
y_train = data_train.label
X_test = data_test.new_reviews
y_test = data_test.copy().label
```

Utilisation du modèle random forest pour apprendre le modèle permettant de détecter le caractère positif ou négatif de la critique. On construit 2 pipelines (un pour chaque encodeur de texte).

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error
pipel = Pipeline([("count", CountVectorizer()), ("rf", RandomForestClassifier(n_estimators = 100))])
pipel.fit(X_train, y_train)
y_pred = pipel.predict(X_test)
y_pred
print(mean_squared_error(y_test, y_pred).sum())
```

```python
pipel2 = Pipeline([("tfidf", TfidfVectorizer()), ("rf", RandomForestClassifier(n_estimators = 100))])
pipel2.fit(X_train, y_train)
y_pred2 = pipel2.predict(X_test)
y_pred2
print(mean_squared_error(y_test, y_pred2).sum())
```

On choisit le pipeline 2 et on calcule le features importance du modèle appris.

```python
imp =pd.Series(pipel2.steps[1][1].feature_importances_ ,index=pipel2.steps[0][1].get_feature_names()).sort_values(ascending = False)
imp.head(20)
```
